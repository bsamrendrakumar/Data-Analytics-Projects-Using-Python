```python
import pandas as pd
import matplotlib.pyplot as plt

# Suppose your dataset has these columns:
# Country | Year | CO2_Emissions_tons_per_capita | Sea_Level_Rise_mm | Avg_Temperature_degC

# Filter top 5 countries
top_countries = ['United States', 'China', 'India', 'Russia', 'Brazil']
df_top = df[df['Country'].isin(top_countries)]

# Create plot
fig, ax1 = plt.subplots(figsize=(12,7))

# Plot CO Emissions
for country in top_countries:
    data = df_top[df_top['Country'] == country]
    ax1.plot(data['Year'], data['CO2_Emissions_tons_per_capita'], marker='o', label=f'{country} - CO')

ax1.set_xlabel('Year')
ax1.set_ylabel('CO Emissions (tons/person)', color='r')
ax1.tick_params(axis='y', labelcolor='r')

# Second Y-axis for Sea Level
ax2 = ax1.twinx()
for country in top_countries:
    data = df_top[df_top['Country'] == country]
    ax2.plot(data['Year'], data['Sea_Level_Rise_mm'], linestyle='--', label=f'{country} - Sea Level')

ax2.set_ylabel('Sea Level Rise (mm)', color='b')
ax2.tick_params(axis='y', labelcolor='b')

# Optional: Add Average Temperature trend (as dotted lines)
for country in top_countries:
    data = df_top[df_top['Country'] == country]
    ax1.plot(data['Year'], data['Avg_Temperature_degC'], linestyle=':', label=f'{country} - Temp (C)')

# Combine legends
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='upper left', fontsize=8)

plt.title('Top 5 CO Emitting Countries: CO, Sea Level & Temperature Trends (20002024)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
